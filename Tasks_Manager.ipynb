```python
# 1- Add tasks to a list
# 2- Mark task as complete
# 3- View tasks
# 4- Quit

# Initialize an empty list to store tasks
tasks = []

# Define the main function
def main():
    # Display the menu options
    message = """1- Add tasks to a list
2- Mark task as complete
3- View tasks
4- Quit"""

    while True:
        # Print the menu and get the user's choice
        print(message)
        choice = input("Enter your choice: ")

        if choice == "1":
            add_task()
        elif choice == "2":
            mark_task()
        elif choice == "3":
            view_tasks(tasks)
        elif choice == "4":
            break
        else:
            print("Invalid choice, please enter a number between 1 to 4")

# Function to add a task
def add_task():
    task = input("Enter your task:")
    task_info = {"task": task, "completed": False}
    tasks.append(task_info)
    print("Task added to the list.")
    print("*" * 30)

# Function to mark a task as complete
def mark_task():
    incomplete_tasks = [task for task in tasks if not task["completed"]]

    if len(incomplete_tasks) == 0:
        print("No task to mark as complete")
        return

    print("Incomplete tasks:")

    for index, task in enumerate(incomplete_tasks, 1):
        print(f"{index}. {task['task']}")

    try:
        task_number = int(input("Please enter the task number to mark as complete: "))

        if task_number < 1 or task_number > len(incomplete_tasks):

          # raise IndexError("Invalid task number") #for programmer
          print("Invalid task number")


        incomplete_tasks[task_number - 1]["completed"] = True
        print("Task marked as complete.")

    except Exception as e:
        print(f"An error occurred: {e}")


# Function to view tasks
def view_tasks(tasks):
    if tasks:
        print("All tasks:")

        for index, task in enumerate(tasks, 1):
            status = "âœ”" if task["completed"] else "ðŸ›¶"
            print(f"{index}. {task['task']} - {status}")
            print("*" * 30)
    else:
        print("No tasks to show")

# Entry point of the script
if __name__ == "__main__":
    main()
    print("This script is the main program.")
# print (type(__name__))               str




```
